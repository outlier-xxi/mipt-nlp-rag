# FinRAD — Financial Readability Assessment Dataset

[FinRAD dataset](https://huggingface.co/datasets/sohomghosh/FinRAD_Financial_Readability_Assessment_Dataset) dataset from Hugging Face (13k+ definitions of financial terms with readability labels)

```python
from datasets import load_dataset
import pandas as pd
```

## Load dataset

```python
finrad = load_dataset("sohomghosh/FinRAD_Financial_Readability_Assessment_Dataset")
finrad
```

## Train split

```python
train = finrad["train"]
print(f"Rows: {len(train)}")
print(f"Columns: {train.column_names}")
train
```

## Train data

```python
df = train.to_pandas()
print(df.shape)
df.head(10)
```

## Main columns

- **terms**: financial term
- **definitions**: definition text
- **source**: source of the term (e.g. prin, investopedia)
- **assigned_readability**: 0 = not readable, 1 = readable
- Other columns: readability scores (flesch_reading_ease, flesch_kincaid_grade, etc.)

```python
df[["terms", "definitions", "source", "assigned_readability"]].head(10)
```

## Label and source distribution

```python
print("assigned_readability value counts:")
print(df["assigned_readability"].value_counts().sort_index())
print()
print("source value counts:")
print(df["source"].value_counts())
```

## Readability score columns (summary)

```python
score_cols = [
    "flesch_reading_ease", "flesch_kincaid_grade", "smog_index",
    "coleman_liau_index", "automated_readability_index", "dale_chall_readability_score",
    "linsear_write_formula", "gunning_fog"
]
df[score_cols].describe()
```
